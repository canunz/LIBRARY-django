```python
# 1. Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 2. Cargar los datos
df = pd.read_csv('/content/drive/MyDrive/Maching learning/Regresión: Valor numérico/datos_exportados (1).csv', sep=',')
```

```python
# 3. Definir las variables
columnas_predictoras = [
    'MatchKills',
    'MatchHeadshots',
    'TravelledDistance',
    'RoundStartingEquipmentValue',
    'TeamStartingEquipmentValue'
]

X = df[columnas_predictoras]
y = df['Survived']
```

```python
# 4. Separar en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 5. Crear y entrenar el modelo Random Forest
modelo_rf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
modelo_rf.fit(X_train, y_train)
```

```python
# 6. Predecir
y_pred = modelo_rf.predict(X_test)
```

```python
# 7. Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión (accuracy):", accuracy)
```

🧮 Precisión del modelo (Accuracy)
📌 Accuracy = 0.6266
👉 Esto significa que el modelo acierta en el 62.7% de los casos.
Es el modelo más preciso hasta ahora, superando a:

Regresión logística (~61%)

Árbol de decisión (~62.2%)

```python
# 8. Matriz de Confusión
matriz = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=matriz, display_labels=modelo_rf.classes_)
disp.plot(cmap='YlGn')
plt.title('Matriz de Confusión - Random Forest')
plt.show()
```

3760 veces acertó al decir que no sobrevivía.

1803 veces acertó al decir que sí sobrevivía.

Se equivocó 1244 veces diciendo que sobrevivía cuando no.

Se equivocó 2071 veces diciendo que no sobrevivía cuando sí.

**Conclusión**

"El modelo de Random Forest fue el que obtuvo mejor precisión, con un 62.7% de aciertos.
Aunque no es perfecto, es mejor que la Regresión Logística y el Árbol de Decisión.
Además, fue más equilibrado en la predicción de jugadores que sí sobreviven, que era uno de nuestros objetivos clave.
Esto demuestra que Random Forest es más confiable para este tipo de predicción, ya que considera múltiples árboles y reduce los errores que se cometen con un único árbol."
