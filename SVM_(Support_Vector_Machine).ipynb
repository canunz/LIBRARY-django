```python
# 1. Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
# 2. Cargar los datos
df = pd.read_csv('/content/drive/MyDrive/Maching learning/Regresión: Valor numérico/datos_exportados (1).csv', sep=',')
```

```python
# 3. Seleccionar variables predictoras y variable objetivo
columnas_predictoras = [
    'MatchKills',
    'MatchHeadshots',
    'TravelledDistance',
    'RoundStartingEquipmentValue',
    'TeamStartingEquipmentValue'
]

X = df[columnas_predictoras]
y = df['Survived']
```

```python
# 4. Separar datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 5. Crear y entrenar el modelo SVM
modelo_svm = SVC(kernel='linear')  # También puedes probar 'rbf'
modelo_svm.fit(X_train, y_train)
```

```python
# 6. Predecir
y_pred = modelo_svm.predict(X_test)
```

```python
# 7. Evaluar
accuracy = accuracy_score(y_test, y_pred)
print("Precisión (accuracy):", accuracy)
```

```python
# 8. Matriz de Confusión
matriz = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=matriz, display_labels=modelo_svm.classes_)
disp.plot(cmap='Oranges')
plt.title('Matriz de Confusión - SVM')
plt.show()
```

📊 Modelos de Clasificación (ordenados por desempeño)
🔻 KNN (K-Nearest Neighbors) — 57.75%

Modelo basado en jugadores “similares”. Su desempeño fue el más bajo, indicando que no logra distinguir bien los casos de supervivencia en este contexto.

🔸 SVM (Support Vector Machine) — 60.63%

Buen modelo en teoría, pero aquí no logró separar bien las clases. Fue mejor que KNN, pero aún por debajo del promedio.

🟡 Regresión Logística — 61.05%

Sencillo, lineal y útil como punto de partida. Su precisión fue decente, pero no superó a los modelos basados en árboles.

🟢 Random Forest — 62.66%

Modelo muy robusto. Combina muchos árboles y tiene buen rendimiento general. Estuvo muy cerca del primer lugar.

🥇 Árbol de Decisión — 62.81%

El modelo más preciso. Además de ser interpretable, superó por poco a Random Forest, lo que lo posiciona como el mejor clasificador para este problema.
